# Импорты

```python
import plotly.express as px
import pandas as pd
import numpy as np
```

# Входные данные

Уравнение:<br>
$
\begin{cases} \displaystyle
    u'' - \frac{3}{(x + 1)^2} u = \frac{-1,5}{\sqrt{x + 1}} \\
    3u(0) - u'(0) = 1\\
    u'(1) = \sqrt{2}
\end{cases}
$

Решение:<br>
$\displaystyle
u(x) = \frac{2}{3} (x + 1)^{\frac{3}{2}}$

```python
N = 10

def p(x: float) -> float:
    return 0

def q(x: float) -> float:
    return -3 / ((x + 1) ** 2)

def f(x: float) -> float:
    return -1.5 / ((x + 1) ** 0.5)

def u(x: float) -> float:
    return 2 / 3 * (x + 1) ** (3 / 2)
```

# График решения

```python
x = np.linspace(0, 1, 500)
u_vals = u(x)

df = pd.DataFrame({
    'x': x,
    'u': u_vals
})

fig = px.line(
    df, x='x', y='u', 
    labels={'x': 'x', 'u': 'u'}
)

fig.update_traces(line=dict(color='blue', width=2))
fig.update_layout(
    template='plotly_white',
    hovermode='x unified'
)

fig.show()
```

# Монотонная прогонка

```python
def monotonous_reduction(A: np.array, B: np.array, C: np.array, F: np.array) -> np.array:
    n = len(A) - 1

    alpha = np.zeros(n + 1)
    beta = np.zeros(n + 1)
    U = np.zeros(n + 1)

    alpha[1] = -C[0] / B[0]
    beta[1] = F[0] / B[0]

    for i in range(1, n):
        alpha[i + 1] = -C[i] / (B[i] + alpha[i] * A[i])
        beta[i + 1] = (F[i] - beta[i] * A[i]) / (B[i] + alpha[i] * A[i])

    U[n] = (F[n] - beta[n] * A[n]) / (B[n] + alpha[n] * A[n])

    for i in range(n - 1, -1, -1):
        U[i] = alpha[i + 1] * U[i + 1] + beta[i + 1]

    return U
```

# Определение системы уравнений

$\displaystyle
p_i = p(x_i), ~~ p(x) = 0 \\
q_i = q(x_i), ~~ q(x) = -\frac{3}{(x + 1)^2} \\
f_i = f(x_i), ~~ f(x) = \frac{-1,5}{\sqrt{x + 1}} \\
x_i = ih \\
h = \frac{1}{N}
$

Система имеет вид:<br>
$\displaystyle
a_i y_{i - 1} + b_i y_i + c_i y_{i + 1} = f_i, ~~ i = \overline{1, N - 1}
$

Так как $\beta_0 = -1 \neq 0$, то необходимо необходимо вывести нулевое уравнение из соотношения:<br>
$\displaystyle
u'(0) \approx \frac{-y_2 + 4y_1 - 3y_0}{2h} \\
3u(0) - 1 = 3y_0 - 1 \approx \frac{-y_2 + 4y_1 - 3y_0}{2h} \\
6h y_0 - 2h = -y_2 + 4y_1 - 3y_0 \\
(6h + 3)y_0 = 4y_1 - y_2 + 2h \\
y_0 = \frac{4y_1 - y_2 + 2h}{6h + 3}
$

1-ое уравнение:<br>
$\displaystyle
a_1 y_0 + b_1 y_1 + c_1 y_2 = f_1 \\
a_1 \frac{4y_1 - y_2 + 2h}{6h + 3} + b_1 y_1 + c_1 y_2 = f_1 \\
\left( \frac{4a_1}{6h + 3} + b_1 \right) y_1 + \left( c_1 - \frac{a_1}{6h + 3} \right) y_2 = f_1 - \frac{2h a_1}{6h + 3} 
$

Так как $\beta_1 = 1 \neq 0$, то необходимо необходимо вывести $N$-ое уравнение из соотношения:<br>
$\displaystyle
u'(1) \approx \frac{3y_N - 4y_{N - 1} + y_{N - 2}}{2h} \\
\sqrt{1} \approx \frac{3y_N - 4y_{N - 1} + y_{N - 2}}{2h} \\
2\sqrt{2} h = 3y_N - 4y_{N - 1} + y_{N - 2} \\
3y_N = -y_{N - 2} + 4y_{N - 1} + 2\sqrt{2} h \\
y_N = -\frac{y_{N - 2} + 4y_{N - 1} + 2\sqrt{2} h}{3} \\
$

$(N - 1)$-ое уравнение:<br>
$\displaystyle
a_{N - 1} y_{N - 2} + b_{N - 1} y_{N - 1} + c_{N - 1} y_N = f_{N - 1} \\
a_{N - 1} y_{N - 2} + b_{N - 1} y_{N - 1} + c_{N - 1} \frac{-y_{N - 2} + 4y_{N - 1} + 2\sqrt{2} h}{3} = f_{N - 1} \\
\left( a_{N - 1} - \frac{c_{N - 1}}{3} \right) y_{N - 2} + \left( b_{N - 1} + \frac{4c_{N - 1}}{3} \right) y_{N - 1} = f_{N - 1} - \frac{2\sqrt{2} h c_{N - 1}}{3}
$

Коэффициенты определяются следующим образом:<br>
$\displaystyle
a_0 = c_N = 0 \\
r = \frac{p_i}{2} h = 0 \\
a_i = \frac{1}{h^2} (1 + 2|r| - 2\sin 2|r| - r) = \frac{1}{h^2}, ~~ i = \overline{1, N - 1} \\
c_i = \frac{1}{h^2} (1 + 2|r| - 2\sin 2|r| + r) = \frac{1}{h^2}, ~~ i = \overline{1, N - 1} \\
b_i = q_i - a_i - c_i = q_i - 2 \frac{1}{h^2}, ~~ i = \overline{1, N - 1}
$

Тогда система имеет вид:<br>
$\displaystyle
\left( \begin{array}{ccccccccc}
    \displaystyle \frac{4a_1}{6h + 3} + b_1 & \displaystyle c_1 - \frac{a_1}{6h + 3} & 0 & 0 & \ldots & 0 & 0 & 0 & 0\\
    a_2 & b_2 & c_2 & 0 & \ldots & 0 & 0 & 0 & 0 \\
    0 & a_3 & b_3 & c_3 & \ldots & 0 & 0 & 0 & 0 \\
    \vdots & \vdots & \vdots & \vdots & \ddots & \vdots & \vdots & \vdots & \vdots \\
    0 & 0 & 0 & 0 & \ldots & a_{N - 3} & b_{N - 3} & c_{N - 3} & 0 \\
    0 & 0 & 0 & 0 & \ldots & 0 & a_{N - 2} & b_{N - 2} & c_{N - 2} \\
    0 & 0 & 0 & 0 & \ldots & 0 & 0 & \displaystyle a_{N - 1} - \frac{c_{N - 1}}{3} & \displaystyle b_{N - 1} + \frac{4c_{N - 1}}{3}
\end{array} \right) \cdot
\left( \begin{array}{c}
    y_1 \\
    y_2 \\
    y_3 \\
    \vdots \\
    y_{N - 3} \\
    y_{N - 2} \\
    y_{N - 1} \\
\end{array} \right) =
\left( \begin{array}{c}
    \displaystyle f_1 - \frac{2h a_1}{6h + 3}  \\
    f_2 \\
    f_3 \\
    \vdots \\
    f_{N - 3} \\
    f_{N - 2} \\
    \displaystyle f_{N - 1} - \frac{2\sqrt{2} h c_{N - 1}}{3}
\end{array} \right)
$

```python
h = 1 / N
x_i = np.array([i * h for i in range(N + 1)])

def r(i: int) -> float:
    return p(x_i[i]) / 2 * h

a_i = np.zeros(shape=(N - 1,), dtype=np.float64)
b_i = np.zeros(shape=(N - 1,), dtype=np.float64)
c_i = np.zeros(shape=(N - 1,), dtype=np.float64)
f_i = np.zeros(shape=(N - 1,), dtype=np.float64)

for i in range(N - 1):
    r_i = r(i + 1)
    
    a_i[i] = 1 / h ** 2 * (1 + 2 * np.abs(r_i) - 2 * np.sin(2 * np.abs(r_i)) - r_i)
    c_i[i] = 1 / h ** 2 * (1 - 2 * np.abs(r_i) + 2 * np.sin(2 * np.abs(r_i)) + r_i)
    b_i[i] = q(x_i[i + 1]) - a_i[i] - c_i[i]
    f_i[i] = f(x_i[i + 1])

f_i[0] = f_i[0] - 2 * h * a_i[0] / (6 * h + 3)
c_i[0] = c_i[0] - a_i[0] / (6 * h + 3)
b_i[0] = 4 * a_i[0] / (6 * h + 3) + b_i[0]
a_i[0] = 0

f_i[N - 2] = f_i[N - 2] - 2 ** 1.5 * h * c_i[N - 2] / 3
a_i[N - 2] = a_i[N - 2] - c_i[N - 2] / 3 
b_i[N - 2] = b_i[N - 2] + 4 * c_i[N - 2] / 3
c_i[N - 2] = 0
```

# Решение системы

```python
y_i = np.zeros(shape=(N + 1,), dtype=np.float64)
y_i[1:-1] = monotonous_reduction(a_i, b_i, c_i, f_i)
y_i[0] = (4 * y_i[1] - y_i[2] + 2 * h) / (6 * h + 3)
y_i[N] = (-y_i[N - 2] + 4 * y_i[N - 1] + 2 ** 1.5 * h) / 3

u_i = np.array([u(x_i[i]) for i in range(N + 1)])

figure = px.line(x=x_i, y=u_i, markers=True)

figure.update_traces(name="u_h", showlegend=True)
figure.add_scatter(x=x_i, y=y_i, name="y_h")

figure.show()
```

# Погрешность

```python
figure = px.line(x=x_i, y=np.abs(u_i - y_i), markers=True)
figure.update_traces(name="|u_h - y_h|", showlegend=True)
figure.update_layout(yaxis=dict(exponentformat="power"))
figure.show()
```
