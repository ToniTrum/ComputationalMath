# Импорты

```python
import numpy as np
import plotly.express as px
```

# Входные данные

Уравнение: <br>
$\displaystyle
u(x) - \int_{0}^{2\pi} |\pi - s| \sin x u(s) ds = x
$

Решение: <br>
$
u(x) = x + \pi^3 \sin x
$

```python
n = 3

def u(x: float) -> float:
    return x + np.pi ** 3 * np.sin(x)
```

# Решение

Для метода Бубнова-Галёркина вначале выбирается линейной независимый базис $\{\varphi_i\}_{0}^{\infty}$, полный в $L_2(0, 2\pi)$, выберем следующий ортогональный базис: <br>
$\displaystyle
\varphi_0 = 1 \\
\varphi_{2i - 1} = \sin(ix), ~~ i = \overline{1, \infty} \\
\varphi_{2i} = \cos(ix), ~~ i = \overline{1, \infty}
$

Решение строится из приближения: <br>
$\displaystyle
u_n(x) = \sum_{i = 0}^{n} C_i \varphi_i(x)
$

Для нахождения коэффициентов $C_i$ решается система: <br>
$\displaystyle
\sum_{k = 0}^n C_k (\varphi_k(x), \varphi_i(x)) - (x, \varphi_i(x)) - \left( \int_{0}^{2\pi} |\pi - s| \sin x \sum_{j = 0}^{n} C_j \varphi_j(s) ds, \varphi_i(x) \right) = 0 \\
(\varphi_k(x), \varphi_i(x)) = \begin{cases}
    (\varphi_0(x), \varphi_0(x)), & k = i = 0 \\
    (\varphi_{2m}(x), \varphi_{2m}(x)), & k = i = 2m, m \in \mathbb{N} \\
    (\varphi_{2m - 1}(x), \varphi_{2m - 1}(x)), & k = i = 2m - 1, m \in \mathbb{N} \\
    (\varphi_0(x), \varphi_{2j}(x)), & k = 0, i = 2j, j \in \mathbb{N} \\
    (\varphi_0(x), \varphi_{2j - 1}(x)), & k = 0, i = 2j - 1, j \in \mathbb{N} \\
    (\varphi_{2m - 1}(x), \varphi_{2j - 1}(x)), & k = 2m - 1, i = 2j - 1, m, j \in \mathbb{N}, m \neq j \\
    (\varphi_{2m}(x), \varphi_{2j}(x)), & k = 2m, i = 2j, m, j \in \mathbb{N}, m \neq j \\
    (\varphi_{2m - 1}(x), \varphi_{2j}(x)), & k = 2m - 1, i = 2j, m, j \in \mathbb{N}
\end{cases} = \begin{cases}
    \displaystyle \int_{0}^{2\pi} dx, & k = i = 0 \\
    \displaystyle \int_{0}^{2\pi} \cos^2(mx) dx, & k = i = 2m, m \in \mathbb{N} \\
    \displaystyle \int_{0}^{2\pi} \sin^2(mx) dx, & k = i = 2m - 1, m \in \mathbb{N} \\
    \displaystyle \int_{0}^{2\pi} \cos(jx) dx, & k = 0, i = 2j, j \in \mathbb{N} \\
    \displaystyle \int_{0}^{2\pi} \sin(jx) dx, & k = 0, i = 2j - 1, j \in \mathbb{N} \\
    \displaystyle \int_{0}^{2\pi} \sin(mx) \sin(jx) dx, & k = 2m - 1, i = 2j - 1, m, j \in \mathbb{N}, m \neq j \\
    \displaystyle \int_{0}^{2\pi} \cos(mx) \cos(jx) dx, & k = 2m, i = 2j, m, j \in \mathbb{N}, m \neq j \\
    \displaystyle \int_{0}^{2\pi} \sin(mx) \cos(jx) dx, & k = 2m - 1, i = 2j, m, j \in \mathbb{N}
\end{cases} = \begin{cases}
    2\pi, & k = i = 0 \\
    \pi, & k = i > 0 \\
    0, & k \neq i
\end{cases} \\
(x, \varphi_i(x)) = \begin{cases}
    (x, \varphi_0(x)), & i = 0 \\
    (x, \varphi_{2j}(x)), & i = 2j, j \in \mathbb{N} \\
    (x, \varphi_{2j - 1}(x)), & i = 2j - 1, j \in \mathbb{N}
\end{cases} = \begin{cases}
    \displaystyle \int_{0}^{2\pi} x dx, & i = 0 \\
    \displaystyle \int_{0}^{2\pi} x \cos(jx) dx, & i = 2j, j \in \mathbb{N} \\
    \displaystyle \int_{0}^{2\pi} x \sin(jx) dx, & i = 2j - 1, j \in \mathbb{N} \\
\end{cases} = \begin{cases}
    2\pi^2, & i = 0 \\
    \displaystyle -\frac{2\pi}{j}, & i = 2j - 1, j \in \mathbb{N} \\
    0, & i = 2j, j \in \mathbb{N} \\
\end{cases} \\
(\sin x, \varphi_i(x)) = \begin{cases}
    \pi, & i = 1 \\
    0, & i \neq 1
\end{cases} \\
\int_{0}^{2\pi} |\pi - s| \sum_{j = 0}^{n} C_j \varphi_j(s) ds = \sum_{j = 0}^n C_j \int_{0}^{2\pi} \left( \int_{0}^{\pi} (\pi - s) \varphi_j(s) ds + \int_{0}^{2\pi} (s - \pi) \varphi_j(s) ds \right) = \pi^2 C_0 + \sum_{j = 1}^{[\frac{n}{2}]} \frac{2((-1)^{k + 1} + 1)}{k^2} C_{2j} = \pi^2 C_0 + \sum_{j = 1}^{[\frac{n}{4}]} \frac{4 C_{4j - 2}}{(2j - 1)^2}
$

Получаем следующую систему: <br>
$\displaystyle
\begin{cases}
    2\pi C_0 - 2\pi^2 = 0 \\
    \displaystyle \pi C_1 + 2\pi - \pi^3 C_0 - 4\pi \sum_{j = 1}^{[\frac{n}{4}]} \frac{C_{4j - 2}}{(2j - 1)^2} = 0 \\
    \displaystyle \pi C_i + \delta_{1}^{i \mod 2} \frac{2\pi}{i} = 0, ~~ i = \overline{2, n}
\end{cases} = \begin{cases}
    C_0 = \pi \\
    C_1 = \pi^3 - 2 \\
    \displaystyle C_{2j + 1} = -\frac{2\pi}{i}, & i = 2k + 1, k = \overline{1, \displaystyle \left[ \frac{n}{2} \right]} \\
    \displaystyle C_{2j} = 0, & i = 2k, k = \overline{1, \displaystyle \left[ \frac{n}{2} \right]} \\
\end{cases}
$

```python
def phi(x: float, i: int) -> float:
    if i == 0:
        return 1.0
    if i % 2 == 0:
        k = i // 2
        return np.cos(k * x)
    else:
        k = (i + 1) // 2
        return np.sin(k * x)

C = [0.0] * (n + 1)
C[0] = np.pi 
C[1] = np.pi ** 3 - 2
for i in range(2, n + 1):
    if i % 2 == 1:
        k = (i + 1) // 2
        C[i] = -2 / k

def u_n(x: float) -> float:
    return sum(C[i] * phi(x, i) for i in range(0, n + 1))
```

# Визуализация

```python
x = np.linspace(0 + 0.1, 2 * np.pi - 0.1, 1000)
u_exact = np.array([u(x_i) for x_i in x])
u_numeric = np.array([u_n(x_i) for x_i in x])
```

## Сравнение

```python
figure = px.line(x=x, y=u_exact, markers=False)
figure.update_traces(name="Точное решение", showlegend=True)

figure.add_scatter(x=x, y=u_numeric, name="Численное решение")

figure.show()
```

## Погрешность

```python
error = [np.abs(u(x_i) - u_n(x_i)) for x_i in x]

figure = px.line(x=x, y=error, markers=False)
figure.update_layout(yaxis=dict(exponentformat="power"))
figure.show()
```
