# Импорты

```python
import numpy as np
import plotly.express as px
```

# Входные данные

Уравнение: <br>
$\displaystyle
u(x) - \int_{0}^{2\pi} |\pi - s| \sin x u(s) ds = x
$

Решение: <br>
$
u(x) = x + \pi^3 \sin x
$

```python
n = 10

def k(x: float, s: float) -> float:
    return np.abs(np.pi - s) * np.sin(x)

def f(x: float) -> float:
    return x

def u(x: float) -> float:
    return x + np.pi ** 3 * np.sin(x)

a = 0
b = np.pi
```

# Решение

Для метода Бубнова-Галёркина вначале выбирается линейной независимый базис $\{\varphi_i\}_{1}^{\infty}$, полный в $L_2[0, 2\pi]$, выберем следующий ортонормированный базис: <br>
$\displaystyle
\varphi_i = \frac{1}{\sqrt{\pi}} \sin(ix), ~~ i = \overline{1, \infty}
$

Решение строится из приближения: <br>
$\displaystyle
u_n(x) = \sum_{i = 1}^{n} C_i \varphi_i(x)
$

Для нахождения коэффициентов $C_i$ решается система: <br>
$\displaystyle
\sum_{k = 1}^n C_k (\varphi_k(x), \varphi_i(x)) - (x, \varphi_i(x)) - \left( \int_{0}^{2\pi} |\pi - s| \sin x \sum_{j = 1}^{n} C_j \varphi_j(s) ds, \varphi_i(x) \right) = 0 \\
(\varphi_k(x), \varphi_i(x)) = \delta_k^i \\
(x, \varphi_i(x)) = \frac{1}{\sqrt{\pi}} \int_0^{2\pi} x \sin(ix) dx = -\frac{2\sqrt{\pi}}{i} \\
\int_{0}^{2\pi} |\pi - s| \sin x \sum_{j = 1}^{n} C_j \varphi_j(s) ds = \frac{\sin x}{\sqrt{\pi}} \sum_{j = 1}^{n} C_j \int_{0}^{2\pi} |\pi - s| \sin(js) ds = \frac{\sin x}{\sqrt{\pi}} \sum_{j = 1}^{n} C_j \left( \int_{0}^{\pi} (\pi - s) \sin(js) ds + \int_{\pi}^{2\pi} (s - \pi) \sin(js) ds \right) = \frac{\sin x}{\sqrt{\pi}} \sum_{j = 1}^{n} C_j \left( -\frac{\pi - s}{j} \cos(js)\Big|_{0}^{\pi} - \frac{1}{j} \int_{0}^{\pi} \cos(js) ds - \frac{s - \pi}{j} \cos(js)\Big|_{\pi}^{2\pi} + \frac{1}{j} \int_{\pi}^{2\pi} \cos(js) ds \right) = \frac{\sin x}{\sqrt{\pi}} \sum_{j = 1}^{n} C_j \left( \frac{\pi}{j} - \frac{1}{j^2} \sin(js) \Big|_{0}^{\pi} - \frac{\pi}{j} + \frac{1}{j^2} \sin(js) \Big|_{\pi}^{2\pi} \right) = 0 \\
\left( \int_{0}^{2\pi} |\pi - s| \sin x \sum_{j = 1}^{n} C_j \varphi_j(s) ds, \varphi_i(x) \right) = 0
$

Получаем следующую систему: <br>
$\displaystyle
C_i = -\frac{2\sqrt{\pi}}{i}, ~~ i = \overline{1, n}
$

```python
C = np.array([-2 * np.sqrt(np.pi) / i for i in range(1, n + 1)])

def phi(x: float, i: int) -> float:
    return 1 / np.sqrt(np.pi) * np.sin(x * i)

def u_n(x: float) -> float:
    return np.sum(C[i - 1] * phi(x, i) for i in range(1, n + 1))
```

# Визуализация

```python
x = np.linspace(a, b + 1, 1000)
u_exact = [u(x_i) for x_i in x]
u_numeric = [u_n(x_i) for x_i in x]
```

## Сравнение

```python
figure = px.line(x=x, y=u_exact, markers=False)
figure.update_traces(name="Точное решение", showlegend=True)

figure.add_scatter(x=x, y=u_numeric, name="Численное решение")

figure.show()
```

## Погрешность

```python
error = [np.abs(u(x_i) - u_n(x_i)) for x_i in x]

figure = px.line(x=x, y=error, markers=False)
figure.update_layout(yaxis=dict(exponentformat="power"))
figure.show()
```
