# Импорты

```python
import plotly.express as px
import pandas as pd
import numpy as np
```

# Входные данные

$\left\{ \begin{array}{l}
    y' = \sin x + y \\
    y(0) = 0
\end{array} \right., ~~ x \in [0; 10], N = 10$

```python
def f(x: float, y: float) -> float:
    return np.sin(x) + y

y_0 = 0

a = 0
b = 10

n = 100
```

# График решения

$\displaystyle y(x) = \frac{e^x - \sin x - \cos x}{2}$

```python
def y(x: float) -> float:
    return (np.exp(x) - np.sin(x) - np.cos(x)) / 2

x = np.linspace(0, 10, 500)
y_vals = y(x)

df = pd.DataFrame({
    'x': x,
    'y': y_vals
})

fig = px.line(
    df, x='x', y='y', 
    labels={'x': 'x', 'y': 'y(x)'}
)

fig.update_traces(line=dict(color='blue', width=2))
fig.update_layout(
    template='plotly_white',
    hovermode='x unified'
)

fig.show()
```

# Определение шага и узлов

$\displaystyle h = \frac{b - a}{N} \\
x_{i + 1} = x_0 + h i, i = \overline{1, n} ~~ x_0 = a$

```python
h = (b - a) / n
x_i = np.linspace(a, b, n + 1)

print(h)
print(x_i)
```

# Метод Эйлера

$y_{i + 1} = y_i + h \cdot f(x_i; y_i)$

```python
y_i_oiler = [y_0]
for i in range(0, n):
    y_i_oiler.append(y_i_oiler[i] + h * f(x_i[i], y_i_oiler[i]))
    
y_i_oiler
```

# Метод Эйлера с пересчётом

$\displaystyle y_{i + 1} = y_i + h \cdot f\left( x_i + \frac{h}{2}; y_i + \frac{h}{2} \cdot f(x_i; y_i) \right)$

```python
y_i_oiler_with_recount = [y_0]
for i in range(0, n):
    y_i_oiler_with_recount.append(
        y_i_oiler_with_recount[i] + h * f(x_i[i] + h / 2, y_i_oiler_with_recount[i] + h / 2 * f(x_i[i], y_i_oiler_with_recount[i]))
    )

y_i_oiler_with_recount
```

# Метод Рунге-Кутта 4-го порядка

$\displaystyle y_{i + 1} = y_i + \frac{h}{6} (k_1 + 2 k_2 + 2 k_3 + k_4) \\
k_1 = f(x_i; y_i) \\
k_2 = f\left( x_i + \frac{h}{2}; y_i + \frac{h}{2} k_1 \right) \\
k_3 = f\left( x_i + \frac{h}{2}; y_i + \frac{h}{2} k_2 \right) \\
k_4 = f(x_i + h; y_i + h k_3)$

```python
y_i_runge_kutta = [y_0]
for i in range(0, n):
    k1 = f(x_i[i], y_i_runge_kutta[i])
    k2 = f(x_i[i] + h / 2, y_i_runge_kutta[i] + h / 2 * k1)
    k3 = f(x_i[i] + h / 2, y_i_runge_kutta[i] + h / 2 * k2)
    k4 = f(x_i[i] + h, y_i_runge_kutta[i] + h * k3)
    y_i_runge_kutta.append(y_i_runge_kutta[i] + h / 6 * (k1 + 2 * k2 + 2 * k3 + k4))

y_i_runge_kutta
```

# Сравнение методов

```python
y_i = [y(x) for x in x_i]

figure = px.line(x=x_i, y=y_i, markers=False)
figure.update_traces(name="y(x)", showlegend=True)

figure.add_scatter(x=x_i, y=y_i_oiler, name="y_h метода Эйлера")
figure.add_scatter(x=x_i, y=y_i_oiler_with_recount, name="y_h метода Эйлера с пересчётом")
figure.add_scatter(x=x_i, y=y_i_runge_kutta, name="y_h метода Рунге-Кутта 4-го порядка")

figure.show()
```

# Погрешности

```python
figure = px.line(
    x=x_i, 
    y=[abs(y(x_i[i]) - y_i_oiler[i]) for i in range(n + 1)], 
    markers=False
)
figure.update_traces(name="R_(y) метода Эйлера", showlegend=True)

figure.add_scatter(
    x=x_i, 
    y=[abs(y(x_i[i]) - y_i_oiler_with_recount[i]) for i in range(n + 1)], 
    name="R_(y) метода Эйлера с пересчётом"
)
figure.add_scatter(
    x=x_i, 
    y=[abs(y(x_i[i]) - y_i_runge_kutta[i]) for i in range(n + 1)], 
    name="R_(y) метода Рунге-Кутта 4-го порядка"
)

figure.show()
```
